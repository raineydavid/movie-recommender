```python
import findspark
findspark.init()

import pyspark
import sys
import re

from pyspark import SparkConf, SparkContext

import numpy as np 
import pandas as pd 

r_cols = ['user_id', 'movie_id', 'rating', 'timestamp']
ratings = pd.read_csv("/homes/sc325/bigdata/ml-latest-small/ratings.csv", sep=',', names = r_cols, usecols=range(4))

m_cols = ['movie_id', 'title', 'genre']
movies = pd.read_csv("/homes/sc325/bigdata/ml-latest-small/movies.csv", sep=',', names = m_cols)

df = pd.merge(ratings, movies)

# drop repeated header
df.drop(df.index[0],inplace=True)

df['rating'] = df['rating'].astype(float)
df['movie_id'] = df['movie_id'].astype(int)
df['user_id'] = df['user_id'].astype(int)

df.set_index(df['user_id'])


```

```python
ratings_matrix = df.pivot_table(rows=['user_id'],cols=['movie_id'],values='rating')
ratings_matrix.fillna( 0, inplace = True )
ratings_matrix.head(15)
print(df.loc[df['user_id'] == 4])
```
